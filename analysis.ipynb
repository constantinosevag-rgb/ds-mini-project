```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("data/raw_data.csv")

# Preview the dataset
print("Dataset preview:")
print(df)
```

```python
# Compute the mean GDP
mean_gdp = df["gdp"].mean()
print(f"Average GDP: {mean_gdp:.2f}")
```

```python
# Create a bar chart of GDP by country
plt.bar(df["country"], df["gdp"])
plt.title("GDP per Country")
plt.xlabel("Country")
plt.ylabel("GDP (USD)")
plt.tight_layout()
plt.savefig("gdp_plot.png")  # saves the chart automatically
plt.show()
```
